```python
from pandas import read_excel
import pandas as pd
```

```python
data = read_excel("input.xlsx", sheet_name="Sheet1")
data = data.dropna(axis=1, how="all")
```

```python
def print_column(d):
    if "informatik" in d.lower():
        return "Informatiker"
    else:
        return "Other"
data["Was machst du beruflich? gehst du noch zur Schule?\n"] = data["Was machst du beruflich? gehst du noch zur Schule?\n"].transform(print_column)
print(data)
```

```python
current_data = data.groupby("Was machst du beruflich? gehst du noch zur Schule?\n")
keys = current_data.groups.keys()

with pd.ExcelWriter("output.xlsx") as writer:
    for i in keys:
        current_data.get_group(i)["Wenn eine Website dich nach Cookie Einstellungen fragt, wählst du einzelne Cookies aus oder akzeptierst du einfach alle?\n"].value_counts().to_excel(writer, sheet_name=str(i))
        data["Wenn eine Website dich nach Cookie Einstellungen fragt, wählst du einzelne Cookies aus oder akzeptierst du einfach alle?\n"].value_counts().to_excel(writer, sheet_name="all")
    
```
